```python
species_count = {'setosa': 0, 'versicolor': 0, 'virginica': 0}

for record in data:
    if record['species'] == 'setosa':
        species_count['setosa'] += 1
    elif record['species'] == 'versicolor':
        species_count['versicolor'] += 1
    else:
        species_count['virginica'] += 1
        
print(species_count)
```

```python
species_count2 = {'setosa': 0, 'versicolor': 0, 'virginica': 0}
for record in data:
    spec=record['species']
    species_count2[spec] += 1
```

```python
species_count2
```

```python
species_count3 = {}
for record in data:
    if record['species'] in species_count3:
        species_count3[record['species']] += 1
    else:
        species_count3[record['species']] = 1
```

```python
def function_makes_tuples(dictionary_list):
    
    tuple_list = [(elem['petalLength'], elem['petalWidth']) for elem in dictionary_list]
    
    return tuple_list
```
